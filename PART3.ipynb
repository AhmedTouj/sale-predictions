```python
import matplotlib.pyplot as plt
```

```python
# Plot histogram
df['Item_MRP'].hist(alpha=0.5, bins=20, label='Item_MRP')
# Customize plot
plt.legend()
plt.title('Histogram of mrp')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Show plot
plt.show()

```

```python
# Plot histogram
df['Item_Outlet_Sales'].hist(alpha=0.5, bins=20, label='Item_Outlet_Sales')
# Customize plot
plt.legend()
plt.title('Histogram of outlet sales')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Show plot
plt.show()
```

```python
series=df['Item_Outlet_Sales']
series.plot.box(title="Box plot of Outlet Sales", xticks=[])
```

```python
df.boxplot(by ='Outlet_Establishment_Year', column =['Item_Outlet_Sales'], grid = False)
```

```python
df.boxplot(by ='Outlet_Location_Type', column =['Item_Outlet_Sales'], grid = False)
```

```python
import seaborn as snr
dfc = df.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64'])
plt.figure(figsize=(10,10))
snr.heatmap(dfc.corr(), cmap='plasma', annot=True)
```
